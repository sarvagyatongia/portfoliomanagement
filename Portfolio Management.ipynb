```python
AAPL['Simple_return'] = (AAPL['Adj Close']/AAPL['Adj Close'].shift(1)) -1
```

```python
print AAPL['Simple_return']
```

```python
AAPL['Simple_return'].plot(figsize =(15,8))
plt.show()
```

```python
AAPL_avg_return_w = AAPL['Simple_return'].mean()
AAPL_avg_return_w
```

```python
AAPL_avg_return_a = AAPL['Simple_return'].mean() * 52
AAPL_avg_return_a
```

```python
print 'Apple average annual return: ' + str(round(AAPL_avg_return_a,5)*100)+ ' %'
```
